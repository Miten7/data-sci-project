```python
df = df.rename(columns={'Reporting_Year / Année':'Reporting_Year','NPRI_ID / No_INRP':'NPRI_ID','Company_Name / Dénomination_sociale_de_l\'entreprise':'Company_Name','Facility_Name / Installation':'Facility_Name','NAICS / Code_SCIAN':'NAICS','NAICS Title / Titre Code_SCIAN':'NAICS Title','PROVINCE':'Province','CAS_Number / No_CAS':'CAS_Number','Substance Name (English) / Nom de substance (Anglais)':'Substance Name','Group (English) / Groupe (Anglais)':'Group','Category (English) / Catégorie (Anglais)':'Category','Quantity / Quantité':'Quantity','Units / Unités':'Units','Estimation_Method / Méthode_d’estimation':'Estimation_Method'})
list(df.columns)
```

```python
# find out the most common unit and convert all others
df['Units'].value_counts()
```

```python
# convert quantity in kg to tonnes
kg_to_tons = 0.001
kg_rows = df['Units'] == 'kg'
df.loc[kg_rows, 'Quantity'] *= kg_to_tons
df.loc[kg_rows, 'Units'] = 'tonnes'
```

```python
# convert quantity in grams to tonnes
g_to_tons = 0.000001
kg_rows = df['Units'] == 'grams'
df.loc[kg_rows, 'Quantity'] *= g_to_tons
df.loc[kg_rows, 'Units'] = 'tonnes'
```

```python
df['Units'].value_counts()
# N/B:
#'g TEQ' - For data in g TEQ, facilities first submit data in grams (g).
# We then apply a toxic equivalency factor (a weighting factor that expresses how toxic each dioxin and furan is compared to the most toxic congener) and convert the reported amount to grams of toxic equivalent (g TEQ)
```

**Extracting the substance dimension**

```python
# get the unique valuse of the substance identifiers (CAS_Number)
unique_cas = df['CAS_Number'].unique()
print(unique_cas)
print(len(unique_cas))
```

```python
# substance_dim_columns = ['CAS_Number','Substance Name', 'Group', 'Category']
substance_dim_columns = ['CAS_Number','Substance Name'] # removed release group and category cause it fits better in release dimension
substance_dim = df[substance_dim_columns].drop_duplicates()
substance_dim = substance_dim.reset_index(drop=True)
```

```python
# generating surrogate keys for substance dimension
substance_dim['Substance_id'] = range(1, len(substance_dim)+1)
# reindex the columns
substance_dim = substance_dim.reindex(columns=['Substance_id'] + list([c for c in substance_dim.columns if c!= 'Substance_id']))
substance_dim
```

```python
substance_dim[substance_dim['Substance Name'] == 'Lead (and its compounds)']
```

```python
substance_dim[substance_dim['CAS_Number'] == 'NA - 17']
```

```python
substance_dim.to_csv('substance_dimension.csv')
```

**Extracting Industry dimension and generating surrogate keys**

```python
df
```

```python
industry_dim_columns = ['NAICS', 'NAICS Title']
industry_dim = df[industry_dim_columns].drop_duplicates('NAICS')
industry_dim = industry_dim.reset_index(drop=True)
industry_dim
```

```python
# generating surrogate keys for industry dimension
industry_dim['Industry_id'] = range(1, len(industry_dim)+1)
# reindex the columns
industry_dim = industry_dim.reindex(columns=['Industry_id'] + list([c for c in industry_dim.columns if c!= 'Industry_id']))
industry_dim
```

```python
industry_dim.to_csv('industry_dimension.csv')
```

**Extracting Release Dimension**

```python
df
```

```python
# release_dim_columns = ['Group','Category']
release_dim_columns = ['Group','Category', 'Reporting_Year']
release_dim = df[release_dim_columns].drop_duplicates()
release_dim = release_dim.reset_index(drop=True)
release_dim
```

```python
# generating surrogate keys for release dimension
release_dim['Release_id'] = range(1, len(release_dim)+1)
# reindex the columns
release_dim = release_dim.reindex(columns=['Release_id'] + list([c for c in release_dim.columns if c!= 'Release_id']))
release_dim.sort_values(by='Group')
```

```python
release_dim.loc[release_dim['Reporting_Year'] == 2021]
```

```python
release_dim.to_csv('release_dimension.csv')
```
