```python
import pandas as pd
```

```python
# load data into dataframe
path = 'NPRI_Geolocations.csv' # path to csv file
df = pd.read_csv(path, low_memory=False)
```

```python
df
```

```python
# just keep the English name for all columns
df.columns = [col.split(' /')[0] for col in df.columns]
list(df.columns)
```

```python
# drop unnecessary
df=df.drop(['Year of last filed report',
            'Physical Land Survey Description',
            'National Topographical Description',
            'Datum',
            'NAICS',
            'Key Industrial Sector Code',
            'Key Industrial Sector (English)',
            'Key Industrial Sector (French)',
            'Census Division (CD) Unique ID',
            'Census Division (CD) Name',
            'Census Sub Division (CSD) Unique ID',
            'Census Sub Division (CSD) Name',
            'Census Metropolitan Area (CMA) Unique ID',
            'Census Metropolitan Area (CMA) Name',
            'Economic Region (ER) Unique ID',
            'Economic Region (ER) Name',
            'Forward Sortation Area (FSA)',
            'Ecozone Unique ID',
            'English Ecozone Name',
            'French Ecozone Name',
            'Unique ID of the Major Drainage Area from the Water Survey of Canada (WSC)',
            'Major Drainage Area English Name',
            'Major Drainage Area French Name',
            'Sulphur Oxide Management Area (SOMA)',
            'Ontario Pollution Emission Management Area (PEMA)',
            'Quebec Pollution Emission Management Area (PEMA)',
            'Quebec City-Windsor Corridor'], axis=1)
list(df.columns)
```

```python
df
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df
```

### Extracting facility dimension

```python
facility_dim_columns = ['NPRI ID', 'Facility Name','City', 'Postal Code', 'Province', 'Longitude', 'Latitude']
facility_dim = df[facility_dim_columns].drop_duplicates(['NPRI ID', 'Facility Name'], keep='first')
facility_dim.reset_index(drop=True, inplace=True)
```

```python
# generating surrogate keys for facility dimension
facility_dim['facility_id'] = range(1, len(facility_dim)+1)
# reindex the columns
facility_dim = facility_dim.reindex(columns=['facility_id'] + list([c for c in facility_dim.columns if c!= 'facility_id']))
facility_dim
```

```python
facility_dim = facility_dim.rename(columns={'NPRI ID': 'NPRI_ID', 'Facility Name': 'Facility_Name'})
facility_dim
```

```python
facility_dim.to_csv('facility_dimension.csv')
```

### Extracting location dimension

```python
location_dim_columns = ['City', 'Province']
location_dim = df[location_dim_columns].drop_duplicates(location_dim_columns, keep='first').dropna(axis=0, how='all')
location_dim.reset_index(drop=True, inplace=True)
```

```python
# generating surrogate keys for location dimension
location_dim['location_id'] = range(1, len(location_dim)+1)
# reindex the columns
location_dim = location_dim.reindex(columns=['location_id'] + list([c for c in location_dim.columns if c!= 'location_id']))
location_dim
```

```python
location_dim.to_csv('location_dimension.csv')
```
